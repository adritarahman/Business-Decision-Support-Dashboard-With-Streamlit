### Calculate current stock

```python
def get_current_stock(product_id):
    quantity_purchased=purchase_df[purchase_df['product_id']==product_id]['quantity_purchased'].sum()
    quantity_sold=sales_df[sales_df['product_id']==product_id]['quantity_sold'].sum()
    current_stock=quantity_purchased-quantity_sold
    return current_stock
```

```python
print(get_current_stock(1095))
```

```python
product_df['current_stock']=product_df['product_id'].apply(
    lambda product_id:get_current_stock(product_id)
)
```

### Get Profit

```python
def get_profit(product_id):
    profit_per_product=product_df[product_df['product_id']==product_id]['selling_price']-product_df[product_df['product_id']==product_id]['cost_price']
    profit_per_product=profit_per_product.iloc[0]
    total_purchased_product=sales_df[sales_df['product_id']==product_id]['quantity_sold'].sum()
    total_profit=profit_per_product*total_purchased_product
    return total_profit
    
```

```python
get_profit(1001)
```

```python
product_df['total_profit']=product_df['product_id'].apply(
    lambda product_id: get_profit(product_id)
)

product_df.head()
```

### Flag Overstocked And Understocked

```python
def get_stock_status(product_id):
    product=product_df[product_df['product_id']==product_id].iloc[0]
    stock=product['current_stock']
    reorder=product['reorder_level']
    if stock<reorder:
        return 'Understocked'
    elif stock>reorder*15:
        return 'Overstocked'
    else:
        return 'Properly stocked'
    
    
        
```

```python
product_df['stock_status']=product_df['product_id'].apply(
    lambda product_id:get_stock_status(product_id)
)
```

```python
product_df['stock_status'].value_counts()
```

### Get revenue per product

```python
def get_revenue(product_df,sales_df,product_id):
    selling_price=product_df[product_df['product_id']==product_id]['selling_price'].iloc[0]
    quantity_sold=sales_df[sales_df['product_id']==product_id]['quantity_sold'].sum()
    revenue=selling_price*quantity_sold
    return revenue
```

```python
def get_sales_between_dates(sales_df,start_date,end_date,location):
    return sales_df[(sales_df['sale_date']>=start_date)
        &(sales_df['sale_date']<=end_date)
        &(sales_df['location'].isin(location))
        ]
    
```

```python
def get_product_of_category(product_df,category):
    return product_df[product_df['category'].isin(category)]
```

```python
def get_understocked_product(product_df):
    return product_df[product_df['stock_status']=='Understocked']
```

```python
get_understocked_product(product_df)
len(get_understocked_product(product_df))
```

```python
def get_summary_kpis(product_df,sales_df):
    total_revenue=product_df['product_id'].apply(
        lambda product_id: get_revenue(product_df,sales_df,product_id)
    ).sum()
    total_profit=product_df['total_profit'].sum()
    total_units_sold=sales_df['quantity_sold'].sum()
    total_understocked_products=len(get_understocked_product(product_df))

    return {
        'Total Revenue (K)': int(total_revenue/1e3),
        'Total Profit (K)': int(total_profit/1e3),
        'Total Units sold (K)': int(total_units_sold/1e3),
        'Total Understocked Products': total_understocked_products
    }
```

```python
start_date=datetime.strptime('2024-01-01','%Y-%m-%d').date()
end_date=datetime.strptime('2024-12-31','%Y-%m-%d').date()
selected_location=['Dhaka']
selected_category=['Groceries']
```

```python
filtered_sales=get_sales_between_dates(sales_df,start_date,end_date,selected_location)
filtered_sales.head()
```

```python
filtered_products=get_product_of_category(product_df,selected_category)

filtered_products.head()
```

```python
understocked_products=get_understocked_product(product_df)
understocked_products.head()
```

```python
get_summary_kpis(filtered_products,filtered_sales)
```
